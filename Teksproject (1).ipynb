# Data Cleaning:

````python
df.isna().sum()
````

````python
null_plot=df.isna().sum().sort_values(ascending=False)
sns.barplot(x=null_plot.index,y=null_plot)
plt.xticks(rotation=60)
plt.yticks(rotation=75)
plt.xlabel("Columns")
plt.ylabel("Number of missing values")
plt.suptitle("Distribution of Missing Data",fontsize=14,y=1.03)
plt.title("Count of Null Values", fontsize=12)
plt.show()
````

````python
df['director'].fillna("unknown",inplace=True)
````

````python
df['cast'].fillna('unknown',inplace=True)
````

````python
df.country.mode()[0]
````

````python
df['country'].fillna('unknown',inplace=True)
````

````python
df['description'].fillna('not available',inplace = True)
````

````python
df['date_added'].mode()[0]
````

````python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
````

````python
df.rating.fillna(df['rating'].mode()[0],inplace=True)
````

````python
df['duration'].fillna(df['duration'].mode()[0], inplace=True)
````

# Uni-Variate Analysis :

````python
df.head()
````

````python
#1. How has the number of movies released per year changed over the last 20-30 years?
movies_per_year=df[df['type']=='Movie']['release_year'].value_counts().sort_index()
plt.figure(figsize=(8,3))
sns.lineplot(x=movies_per_year.index,y=movies_per_year.values)
plt.title('Number of movies released per year')
plt.xlabel('years')
plt.ylabel('Number of movies')
plt.show()
````

````python
#2. Comparison of tv shows vs. movies.
sns.countplot(data=df,x='type',hue='type')
plt.show()
````

````python
#3. What is the best time to launch a TV show?
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
````

````python
#Tv shows
months=df[df['type']=='TV Show']['date_added'].dt.month_name()
count=months.value_counts().sort_values(ascending=False)
sns.barplot(x=count.index,y=count.values, color='skyblue')
plt.xticks(rotation=60)
plt.xlabel('Moths')
plt.ylabel('Count')
plt.show()
````

````python
#4.Which country has the highest number of shows?
country_count=df['country'].value_counts().head(10)
sns.barplot(x=country_count.index,y=country_count.values)
plt.xticks(rotation=60)
plt.show()
````

````python
#5.What is the most common rating on Netflix?
sns.countplot(y=df['rating'], order=df['rating'].value_counts().index,color='skyblue')
````

# Multi-Variate Analysis

````python
def convert_duration(x):
    if pd.isna(x):
        return np.nan
    if 'min' in x:
        return int(x.split()[0])
    elif 'Season' in x:
        return int(x.split()[0]) * 60
    else:
        return np.nan

df['duration_mins'] = df['duration'].apply(convert_duration)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='release_year',y='duration_mins',hue='type',alpha=0.6, s=40)

plt.title("Release Year vs Duration (in minutes) by Type", fontsize=14)
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
````
